We initially scraped the XML files in the movies metadata folder (located at `movies_data`) for all the fields above, using BeautifulSoup. 

```python
# Code to scrape the XML files
def find_list(movie, top_level, second_level=None):
    ''' finds a list of attributes 
    movie: The movie tag from a single movie's XML file
    top_level: Top level nested tag
    second_level: Second level nested tag

    if second_level is None, finds all the tags matching the top_level.
    otherwise finds all the tags matching second_level inside the
    first instance of top_level. 

    Returns a list of unicode strings. If nothing is found, 
    returns an empty list
    '''
    if second_level == None:
        return [unicode(x.string) for x in movie.find_all(top_level)]
    else:
        return [unicode(x.string) for x in 
                movie.find(top_level).find_all(second_level)]

def find_single(movie, tag):
    ''' finds a single instance of a tag's string. 
    If tag not found, returns NaN
    '''
    res = movie.find(tag)
    if res == None:
        return np.nan
    else:
        return unicode(res.string)

singletons = ['name', 'company', 'release_date', 'running_time', 'rating',
'US_Gross', 'production_budget', 'weekend_gross', 'number_of_screens',
'oscar_winning_actors_present', 'num_oscar_winning_actors',
'oscar_winning_directors_present', 'num_oscar_winning_directors',
'summer_release', 'christmas_release', 'memorial_release',
'independence_release', 'labor_release']

both_levels = ['genres', 'authors', 'directors', 'origins', 'actors']

one_level = ['oscar_winning_directors', 'oscar_winning_actors']
def scrape(folder, out="metadata.pkl"):
    ''' scrape the movies metadata into a pandas dataframe.
    pickles this dataframe to disk
    folder: folder containing the movies metadata
    out: output filename for the pickled file
    '''
    data_dict_list = list()
    for fname in os.listdir(folder):
        with open(os.path.join("movies_data", fname)) as f:
            soup = BeautifulSoup(f, "xml")
            movie = soup.movie
            d = dict()

            for field in singletons:
                d[field] = find_single(movie, field)

            for field in both_levels:
                field_singular = field[:-1]
                d[field] = find_list(movie, field, field_singular)

            for field in one_level:
                d[field] = find_list(movie, field)
            
            d['key'] = os.path.splitext(fname)[0]

            data_dict_list.append(d)
            
    metadata = pd.DataFrame(data_dict_list)
    metadata.to_pickle(out)
```

### 1.2: Parsing Academy Award Winners from Wikipedia

We scraped Academy Award Winners from the Wikipedia articles for the 78th to 82nd Academy Awards. The data will be coded in a Pandas dataframe with the following columns:

| Field Name      | Field Type    | Description
|-----------------|---------------|----------------------------------------------
|year             | str           | Year for which the Academy Award was presented
|award            | str           | Name of the Academy award
|movie_nominee    | str           | Name of film nominated
|winner           | int           | 1 if the film won, 0 otherwise

```python
def scrape_oscars():
    years = [2005, 2006, 2007, 2008, 2009]
    url_ths = ['78th', '79th', '80th', '81st', '82nd']

    results = list()

    for year, url_th in itertools.izip(years, url_ths):

        # Find the H3 Awards header
        url = 'http://en.wikipedia.org/wiki/%s_Academy_Awards'%(url_th)
        r = requests.get(url)
        res_soup = BeautifulSoup(r.content)

        awards_h3 = None
        for x in res_soup.find_all("h3"):
            res = x.find("span", text="Awards")
            if res:
                awards_h3 = res
                break

        # Get the next table
        table = awards_h3.find_next("table")

        next_fields = None
        for tr in table.children:
            try:
                th = tr.find_all("th")
            except AttributeError:
                pass
            else:
                if th:
                    next_fields = [unicode(t.find('a').string) for t in th]
            try:
                td = tr.find_all("td")
            except AttributeError:
                pass
            else:
                if td:
                    for (award_ind, t) in enumerate(td):
                        lis = t.find_all("li")
                        for (ind, li) in enumerate(lis):
                            movie = unicode(li.find('i').string)
                            #movie = unicode(li.find('i').find('a').string)
                            winner = 0
                            if ind == 0:
                                winner = 1
                            d = {'year': year, 'award': next_fields[award_ind],
                                 'movie_nominee': movie, 'winner': winner}
                            results.append(d)
    return results

#oscar = pd.DataFrame(scrape_oscars())
```

```python
def transform(movie_title):
    ''' Formats a movie title into the keys for the metadata 
    and reviews dataframe. Removes spaces and converts all
    characters to lowercase.
    movie_title: str; the title of the movie
    '''
    joined = "".join([s.lower() for s in movie_title.split(" ")])
    temp1 = joined.replace(',', '')
    temp2 = temp1.replace(':', '')
    temp3 = temp2.replace('/', '')
    temp4 = temp3.replace('-', '')
    temp5 = temp4.replace("'", '')
    temp6 = temp5.replace('.', '')
    return temp6

oscar['mykey'] = oscar['movie_nominee'].map(transform)
oscar.head()
```

After several attempts, we decided to try matching by movie title. We re-ordered titles such as "The Queen" as "Queen, The" and then performed fuzzy matching on the title strings. This got us really close, and we take care of the remaining troublesome cases manually. 

```python
# Reorder titles
start_tokens = ["An ", "The ", "A "]
def reorder(title):
    for s in start_tokens:
        if title.startswith(s):
            spl = title.split(s)
            return ("".join(spl[1:]) + ", " + s).strip()
    return title
        
oscar['new_title'] = oscar['movie_nominee'].map(reorder)
```

```python
def fuzzy_match(s, to, cutoff = 0.65):
    matches = difflib.get_close_matches(s, to, 1, cutoff)
    if len(matches) > 0:
        return matches[0]
    else:
        return np.nan

oscar['name'] = oscar['new_title'].map(lambda x: fuzzy_match(x, metadata['name']))
```

### 2.1: Classifier Tuning on a Training Set

We want to build a classifier that will put a review into one of two bins: positive, and negative. We thus need a training set. We use a dataset prepared by <a href="http://www.cs.cornell.edu/people/pabo/movie-review-data/">Pang and Lee</a>, consisting of 2000 movie reviews, each of which is manually tagged as either negative or positive (there are 1000 positives and 1000 negatives in the dataset).

```python
# Reading in the training data from Pang and Lees
sentiment_data = sklearn.datasets.load_files("pang_review_data")
```

The multinomial **Naive Bayes** classifier, **linear SVM**, and **KNN** all have enjoyed popularity in the sentiment analysis literature, hence we will test these on our set. Since we are pitting three classifiers against each other, we will want to further split the Pang dataset into two partitions. One (train) will be used to tune each individual classifier via 10-fold cross validation, and the other (test) will be used to assess the accuracy of the three tuned classifiers. Tendentially, we will want to choose the classifier that has the best performance on the test portion of the Pang dataset.

N.B.: **Random forests** would also be a good choice of classifier here, but unfortunately the scikit-learn implementation of the random forest classifier does not yet support sparse representations of the feature matrix. For reasons that will be clear in a moment, we need to use sparse matrices for this classification problem. 

```python
# Train-test split
X_train, X_test, Y_train, Y_test = sklearn.cross_validation.train_test_split(
    sentiment_data.data, sentiment_data.target, test_size=0.33, random_state=0)
```

Before we can use a classifier on the textual data, we need to extract numerical features from it. This is achieved by first pre-processing the text so as to get rid of extremely common and extremely uncommon words. (Sometimes <a href="http://www.nltk.org/api/nltk.stem.html">stemming</a> is also used, although we do not employ a stemmer here.) The document is then represented as a bag of one-word or two-word tokens. Each document is represented as a vector of token frequencies. More specifically, these are **term frequency–inverse document frequencies**. The tf-idf frequency is directly proportional to the frequency of the token in the document, and inversely proportional to the frequency of the token in the whole corpus, so that very common words are not given as much weight in the classification. Because there are many possible tokens, and the tf-idf counts for most of them in a given document will be equal to zero, it is pretty much imperative that we use **sparse matrices**. The pre-processing and vectorization is achieved with the `sklearn` module `TfidfVectorizer()`. In order to automate the process of vectorization and classification, we build `sklearn` pipelines:

```python
# Vectorizer-classifier pipelines
pipelines = {'Bayes' : sklearn.pipeline.Pipeline([('vect', TfidfVectorizer(min_df=2, max_df=0.80)),
                        ('clf', sklearn.naive_bayes.MultinomialNB())]),
             'SVM' : sklearn.pipeline.Pipeline([('vect', TfidfVectorizer(min_df=2, max_df=0.80)),
                        ('clf', sklearn.svm.LinearSVC())]),
             'KNN' : sklearn.pipeline.Pipeline([('vect', TfidfVectorizer(min_df=2, max_df=0.80)),
                        ('clf', sklearn.neighbors.KNeighborsClassifier())])}
```

We set up parameter grids for performing cross-validation on the three given vectorizer-classifier pipelines. For each of the pipelines, we perform a search to see whether one-word tokens or two-word tokens yield better predictive power (this is the `ngram_range` parameter). We also tune the regularization parameter for the SVM classifier, and the number of neighbors for the KNN classifier. 

```python
# Parameter vectors for cross-validation
parameters = {'Bayes' : {'vect__ngram_range': [(1, 1), (1, 2)]},
              'SVM' : {'vect__ngram_range': [(1, 1), (1, 2)],
                  'clf__C': np.logspace(0, 3, 4)},
              'KNN' : {'vect__ngram_range': [(1, 1), (1, 2)],
                  'clf__n_neighbors' : range(1, 52, 10)}}
```

From the scraped metadata dataframe, the reviews dataframe, and the oscar dataframe, we seek to create a merged features matrix. We first clean up the data by converting them to the correct types.
```python
# Some housekeeping devices
def convertInt_NaN(str_int):
    try:
        str_int = str_int.replace(",", "")
        return int(float(str_int))
    except ValueError:
        return np.nan

def convertDollars(str_int):
    try:
        if np.isnan(str_int):
            return np.nan
    except TypeError:
        try:
            return int(float(str_int[1:].replace(",", "")))
        except ValueError:
            return np.nan

def convertInt_0(str_int):
    try:
        return int(str_int)
    except ValueError:
        return 0

def convertBool(str_bool):
    if str_bool == 'false':
        return 0
    else:
        return 1

def convertDate(str_date):
    return dateutil.parser.parse(str_date)

def convertOrigins(str_origins):
    if "USA" in str_origins:
        return 1
    else:
        return 0

def dateToDayIndicator(date):
    year = date.year
    base_date = datetime.date(year, 1, 1)
    delta = datetime.date(date.year, date.month, date.day) - base_date
    return delta.days
```

```python
# Converting metadata columns to appropriate types
for field in ['US_Gross', 'weekend_gross', 'production_budget']:
    metadata[field] = metadata[field].map(convertDollars)
for field in ['christmas_release', 'independence_release', 'labor_release', 'memorial_release', 'summer_release']:
    metadata[field] = metadata[field].map(convertBool)
for field in ['num_oscar_winning_actors', 'num_oscar_winning_directors']:
    metadata[field] = metadata[field].map(convertInt_0)
for field in ['running_time', 'number_of_screens']:
    metadata[field] = metadata[field].map(convertInt_NaN)
for field in ['release_date']:
    metadata[field] = metadata[field].map(convertDate)
    metadata['day_of_year'] = metadata[field].map(dateToDayIndicator)
for field in ['origins']:
    metadata['domestic'] = metadata[field].map(convertOrigins)
```

We run multiple linear regression on the features matrix, regressing the Oscar nominee status of a movie on (a subset of) the features. We see that the log US gross, the day of year the movie was released, the running time of the movie, and the review scores of the movie are all highly correlated with the Oscar nomination status. We further explore these effects marginally. For each of the marginal effects we investigated, we first graphically explored the effect using paired boxplots. We then formally tested the effect using a **rank-sum test**. A rank-sum test is a distribution-free test that tests the equality of medians. We chose the rank-sum test because it does not make any assumptions about the underlying distribution of the data. It just requires independence of samples. Even though a movie can influence another movie's box office performance and review score, we assume that these effects are minimal and thus we do not model them. Additionally, we hypothesize that movie companies do not release "big" movies very close to each other for fear of competition, providing some justification to our assumption.

First we standardize our non-categorical data. This would be especially important in improving the performance of an SVM classifier, which is **not scale-invariant**. That said, standardizing is in general a good practice.

The first two classifier are chosen because of their documented predicted power. The third classifier is chosen because we are dealing with binary outcome, and logistic regresion in this case tends to be a good, simple baseline classifier. We write a function to automate the 10-fold cross-validation procedure: 

```python
# Function to automate cross-validation routine
def cross_validation_routine(classifier, parameters, X, Y):
    '''
    Automate cross validation routine.
    '''
    cross_validator = GridSearchCV(classifier, parameters, cv=10, scoring='f1')
    cross_validator.fit(X, Y)
    return cross_validator
```

### 5.4: Visualizing Decision Boundaries

Lastly, we plot decision boundaries for the various classifiers to get some intuition about their bias and variance, keeping in mind the bias-variance tradeoff. We only consider the top two most important features, as judged by the relative importance of the features, returned by the random forest classifier. 

```python
def plot_decision_surface(clf, X_train, Y_train, title):
    plot_step=0.1
    
    if X_train.shape[1] != 2:
        raise ValueError("X_train should have exactly 2 columnns!")
    
    x_min, x_max = X_train[:, 0].min() - plot_step, X_train[:, 0].max() + plot_step
    y_min, y_max = X_train[:, 1].min() - plot_step, X_train[:, 1].max() + plot_step
    xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                         np.arange(y_min, y_max, plot_step))

    clf.fit(X_train,Y_train)
    if hasattr(clf, 'predict_proba'):
        Z = clf.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:,1]
    else:
        Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])    
    Z = Z.reshape(xx.shape)
    cs = plt.contourf(xx, yy, Z, cmap=plt.cm.Reds)
    plt.scatter(X_train[:,0],X_train[:,1],c=Y_train,cmap=plt.cm.Paired)
    plt.title(title);
    plt.xlabel('Normalized Mean Review Score');
    plt.ylabel('Normalized U.S. Gross (Log USD)');
    plt.show()
```

```python
# Subsetting the X matrix
X_nominee_train_subset = X_nominee_train[:,[11,12]]
```

```python
# Plotting random forest classifier surface
plot_decision_surface(sklearn.ensemble.RandomForestClassifier(random_state=0, 
                     n_estimators=tuned_classifiers_nominee['Forest'].best_params_['n_estimators']), 
                      X_nominee_train_subset, Y_nominee_train,
                      'Decision Surface: Random Forest Classifier')
```

The decision surface for the Random Forest classifier shows us that the classifier has low bias. It classifies almost all of the points correctly. However, it has high variance. The decision surface is highly irregular, suggesting that a small change in the explanatory variables of a point would result in a different classification. 

```python
# Plotting decision tree classifier surface
plot_decision_surface(DecisionTreeClassifier(), 
                      X_nominee_train_subset, Y_nominee_train,
                      'Decision Surface: Decision Tree Classifier')
```

The decision surface for a decision tree, which is used in the random forest classifier, is shown above. This decision surface is highly variable, but not too biased. By aggregating decision trees, the random forest reduces the variability of the decision tree classifier but retains its low bias. 

```python
# Plotting logistic regression classifier surface
plot_decision_surface(sklearn.linear_model.LogisticRegression(random_state=0, class_weight='auto'), 
                      X_nominee_train_subset, Y_nominee_train,
                      'Decision Surface: Logistic Regression Classifier')
```

The logistic regression's decision surface has higher bias, because a lot of examples were classified incorrectly. However, it does have lower variance, in that changing the explanatory variables of the data point a little won't change the probability (of getting nominated for an Oscar) by very much. The decision surface is also much more regular. 

```python
# Plotting SVM classifier surface
plot_decision_surface(sklearn.svm.SVC(C=100, gamma=1.0, probability=True, class_weight='auto'), 
                      X_nominee_train_subset, Y_nominee_train,
                      'Decision Surface: SVM Classifier, RBF Kernel, C=100')
```

```python
plot_decision_surface(sklearn.svm.SVC(C=1, gamma=1.0, probability=True, class_weight='auto'), 
                      X_nominee_train_subset, Y_nominee_train,
                      'Decision Surface: SVM Classifier, RBF Kernel, C=1')
```

The SVM decision surfaces are quite circular, because we chose to use an RBF kernel. The decision surfaces are less variable than those of the random forest, but less than those of the logistic regression. They also display bias in between that of the logistic regression and random forests. If we decrease C, we make the SVM less flexible, so we see higher variance but lower bias.
